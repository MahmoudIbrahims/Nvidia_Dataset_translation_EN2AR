```python
# Assuming ds is your DatasetDict
train_subset = ds['train'].remove_columns(['helpfulness', 'correctness', 'coherence', 'complexity', 'verbosity'])

# Alternatively, directly select the required columns
train_subset = ds['train'].select_columns(['prompt', 'response'])

# Replace the original 'train' with the filtered one if needed
ds['train'] = train_subset
```

```python
from deep_translator import GoogleTranslator
import nltk
import json
from google.colab import files  


nltk.download('punkt')
nltk.download('punkt_tab')

def transArabic(text):
    """
    Translates text to Arabic, handling texts longer than 5000 characters.

    Args:
        text: The input text to be translated.

    Returns:
        The translated Arabic text.
    """
    try:
        if len(text) < 5000:
            return GoogleTranslator(source='en', target='ar').translate(text)
        else:
          
            sentences = nltk.tokenize.sent_tokenize(text)

            translated_sentences = [GoogleTranslator(source='en', target='ar').translate(sentence) for sentence in sentences]
            return ' '.join(translated_sentences)
    except Exception as e:
        print(f"Translation failed: {e}")

        return text

import pandas as pd  


first_100_rows = ds['train'].select(range(15000,20300))  # step1= 0:1000,step2= 1000 :2000 ...


first_100_translated = first_100_rows.map(
    lambda x: {
        'prompt': transArabic(x['prompt']),
        'response': transArabic(x['response']),
    }
)


translated_data = pd.DataFrame(first_100_translated)


output_file = 'first_15k_20k_translated.csv'
translated_data.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"Dataset with first 100 rows saved as {output_file}")

files.download(output_file)
```
